```python
import re
import string
from nltk.tokenize import WhitespaceTokenizer 
```

```python
white_space_tokenizer = WhitespaceTokenizer()
```

```python
print(string.punctuation)
```

```python
punch =  string.punctuation
characters_to_remove = '/&_'
for character in characters_to_remove:
    punch = punch.replace(character,'')
print(punch)
```

```python
def PreprocessString(string_data):
    # Remove punctuations, each punctuation = space, ex: ""information @#$retrieval" -> "information    retrieval"
    preprocess_data = re.sub('[%s]' % re.escape(punch), ' ', string_data)
        
    # Tokenize word by white space
    preprocess_data = white_space_tokenizer.tokenize(preprocess_data)
      
    return preprocess_data
```

```python
train_file = 'data.txt'
sentences = list()
with open(train_file, "r", encoding='utf-8') as f:
    content = f.readlines()
    for i in range(len(content)):
        sentence = content[i]
        sentences.append(sentence)

corpus =   []
for e in sentences:
    corpus.append(PreprocessString(e))

print(corpus)
```

```python

l = [e for e in corpus ]
list_words = []
list_pos = []
for e in l:
    pos = [i for i in e if len(i) ==1 and i.isalpha()]
    words = [i for i in e if len(i) > 1 ]
    list_words.append(words)
    list_pos.append(pos)
```

```python
print(list_words)
print(list_pos)
```

```python
# Tìm ra tất cả các pos có trong data  không trùng nhau 
def findPosAndWords(list_pos,list_words):
    tmp1 = []
    tmp2 = []
    set_pos =set()
    set_words = set()
    # convert all to set 
    for i in list_pos:
        tmp1.append(set(i))
    for i in list_words:
        tmp2.append(set(i))
    # Union all 
    for i in tmp1:
        set_pos = set.union (set_pos,i)
    for i in tmp2:
        set_words = set.union (set_words,i)
    return list(set_words),list(set_pos)
```

```python
set_words, set_pos = findPosAndWords(list_pos,list_words)
print(set_words,set_pos)
```

```python
def init_dicA(set_pos):
    dicA = dict()
    set_pos1  = ['s'] + set_pos 
    for i in set_pos1:
        for j in set_pos:
            key = i+j 
            dicA[key] = 0 
    return dicA
```

```python
dicA = init_dicA(set_pos)
print(dicA)
```

```python
def init_dicB(set_words,set_pos):
    dicB= dict()
    n_words = 0 
    unknowns = []
    for e in list_words:
        n_words+= len(e)
    n_unknowns =  int( n_words/len(list_words) )
    for i in set_pos:
        for j in set_words:
            key = i+j 
            dicB[key] = 0 
    for x in range(n_unknowns):
        unknowns.append('unknown'+str(x))
    for x in set_pos:
        for e in unknowns:
            key = x+e 
            dicB[key] = 0
         
   
    
    return dicB

    
    
```

```python
dicB = init_dicB(set_words,set_pos)
print(dicB)
```

```python
def count_pos_A(list_pos, dicA):
    for i in range(0, len(list_pos)):
        key = 's' + list_pos[i]
        if i == 0: 
            dicA[key] +=1
  
    for i in range(0,len(list_pos)):
        for j in range(i,len(list_pos)):
            key = list_pos[i]+list_pos[j]
            if j == i+1:
                dicA[key] +=1

    return dicA 
            

# [['a', 'a', 'n', 'v', 'a', 'n', 'a'], ['a', 'a', 'n', 'v', 'v', 'a', 'n', 'n']]
```

```python
def matrixA(list_pos,dicA):
    for e in list_pos:
        count_pos_A(e,dicA )
    return dicA
```

```python
A = matrixA (list_pos,dicA)
print(A)
```

```python
def count_pos_B(list_words,list_pos,dicB):
    for (item1,item2) in zip(list_pos,list_words):
        for (item3,item4 )in zip (item1,item2):
            key = item3 + item4 
            dicB[key]+=1 
    return dicB 
```

```python
B = count_pos_B(list_words,list_pos,dicB)
print(B)
```

```python
def smoothing(dic):
    for e in dic.keys():
        dic[e]=  dic[e] +1 
    return dic
```

```python
A_after_smooth = smoothing(A)
print(A_after_smooth)
B_after_smooth = smoothing(B)
print(B_after_smooth)
```

```python
def calculateSumA():
    newdict = A.copy()
    for key, value in A.items() :
        key = key[0]+'sum'
        if key in newdict: 
            newdict[key] += value
        else:
            newdict[key] = value
            
    return newdict
    
```

```python
def calculateSumB():
    newdict = B.copy()
    for key, value in B.items() :
        key = key[0]+'sum'
        if key in newdict: 
            newdict[key] += value
        else:
            newdict[key] = value
            
    return newdict
```

```python
A_after_calculate_sum  = calculateSumA()
print(A_after_calculate_sum)
```

```python
B_after_calculate_sum = calculateSumB()
print(A_after_calculate_sum)
```

```python
def calculatePro(dic):
#     newdict = A.copy()
    for key in dic:
        key_sum = key[0]+'sum'
        dic[key] =  dic[key]/dic[key_sum]
    return dic
    
```

```python
A = calculatePro(A_after_calculate_sum)
print(A)
```
